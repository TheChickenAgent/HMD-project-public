```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Data to plot
labels = 'Yes', 'No'
sizes = [12, 3]
colors = ['green', 'red']

fig, ax = plt.subplots()

# Plotting the pie chart with relative numbers
wedges, texts, autotexts = ax.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%')

# Adding absolute numbers to the pie chart
for i, a in enumerate(autotexts):
    a.set_text(f'{sizes[i]} ({a.get_text()})')

plt.title('Did the system correctly understand you?')
plt.legend()
plt.show()

print(sum(sizes))
```

```python
likert_labels = ['1', '2', '3', '4', '5']
scores = [0, 0, 3, 3, 9]
scores_adjusted = scores/np.sum(scores)*100

fig, ax1 = plt.subplots()

# Plotting the percentages
bars = ax1.bar(likert_labels, scores_adjusted)
ax1.set_xlabel('Likert scale')
ax1.set_ylabel('Percentage')
ax1.tick_params(axis='y')

# Creating a secondary y-axis to plot the absolute numbers
ax2 = ax1.twinx()
bars2 = ax2.bar(likert_labels, scores)
ax2.set_ylabel('Absolute numbers')
ax2.tick_params(axis='y')

plt.title('Would you use this system again?')
plt.show()

print(sum(scores))
mean_score = np.dot(scores, np.arange(1, 6))/np.sum(scores)
print(f"Mean: {np.mean(mean_score)}")
```
